# 11.

```python
#plot-этот атрибут есть у коллонок и у всего датафрейма. Он содержит набор методов, которые позволяют вызвать визуальзацию
#hist-гистограмма распределения
#matplotlib inline-визуализация бцдет отображена внутри ноутбука
#bins-колонки
%matplotlib inline
pr = df.price.plot.hist()
pr.set_title('Price Histogram', fontdict={'fontsize': 20})
pr.figure.savefig('gistogram.png')
```

# 12. 

```python
#kde-плотностное распределение
%matplotlib inline
po = df.points.plot.kde()
po.set_title('Points', fontdict={'fontsize': 20})
po.figure.savefig('points.png')
```
